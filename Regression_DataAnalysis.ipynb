DATA ANALYSIS

```python
dfnew = df
```

```python
dfnew['FLOODS']= np.where(dfnew['FLOODS']=='NO',0,1)
```

```python
dfnew.head()
```

```python
dfnew['FLOODS'].value_counts()
```

```python
dfnew.hist(bins=15, figsize=(20,15))
plt.show()
```

```python
#Constructing pie chart
```

```python
percentage = dfnew['FLOODS'].value_counts(normalize=True)*100
```

```python

classlabels = ["YES", "NO"]
plt.figure(figsize=(5, 7))
plt.pie(percentage, labels=classlabels, autopct="%1.1f%%")
plt.title("PIE CHART OF FLOODS")
plt.show()
```
